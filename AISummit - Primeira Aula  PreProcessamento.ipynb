```
import nltk
from pprint import pprint
nltk.download("all")

clear_output()
```

```
from urllib.request import urlopen
txt = urlopen("http://www.gutenberg.org/files/5805/5805.txt")
with open('test.txt','wb') as output:
  output.write(txt.read())
```

```
f = open("test.txt","r")
raw_undecoded = f.read()
raw_puro = raw_undecoded.lower()

print("Quantidade de caracteres: {}".format(len(raw_puro)))
print("Previa do documento: ")
pprint(raw_puro[:200])
```

# **Tokenização do texto por palavra**

---



```
tokens = nltk.word_tokenize(raw_puro)
pprint(tokens[:30])
```

# Tokenização do texto por sentença

```
sentencas = sent_text = nltk.sent_tokenize(' This is an example sent. The sentence splitter will split on sent markers. Ohh really !!')
pprint(sentencas)
```

# Removendo stop-words.

```
en_stop_words = nltk.corpus.stopwords.words('english')
pprint(en_stop_words[:30])
```

```
stopped_tokens = [] # vai guardar os tokens que não sejam stop words

for palavra in tokens:
  if palavra in en_stop_words:
    continue
  else:
    stopped_tokens.append(palavra)
pprint('Lista tokens sem stop words: ')
pprint(stopped_tokens[:20])
```

# Stemização

```
from nltk.stem.porter import PorterStemmer

p_stemmer = PorterStemmer()
stemmed_tokens = [p_stemmer.stem(token) for token in stopped_tokens]
print('Lista de tokens stemizados:')
pprint(stemmed_tokens[:20])
```

# POS-Tag

Tabela com os significados - https://www.ling.upenn.edu/courses/Fall_2003/ling001/penn_treebank_pos.html

```
tagged_tokens = nltk.pos_tag(stemmed_tokens)
pprint(tagged_tokens[:20]) 

# NN - substantivo
# JJ - adjetivo
```

# Encontrar os bi-grams.


```
from gensim.models.phrases import Phrases
```

```
documents = ["the mayor of new york was there", "new york mayor was present", "eu moro em ribeirao preto", "ribeirao preto e grande", "machine learning can be useful sometimes"]

sentence_stream = [doc.split(" ") for doc in documents]
pprint(sentence_stream)
bigram = Phrases(sentence_stream, min_count=1, threshold=1)
```

```
sent = [u'the', u'mayor', u'of', u'new', u'york', u'was', u'there']
print(bigram[sent])

sent = [u'ribeirao', u'preto', u'eh', u'uma', u'linda', u'cidade']
print(bigram[sent])

sent = [u'i', u'love', u'machine', u'learning']
print(bigram[sent])
```
