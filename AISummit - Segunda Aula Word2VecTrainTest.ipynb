```python
#Bibliotecas necessarias
!pip3 install gensim googledrivedownloader nltk

import multiprocessing
import logging, nltk
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
from gensim.corpora import  WikiCorpus
from gensim.models import Word2Vec
import gensim
from gensim.models.word2vec import LineSentence
nltk.download('punkt')
```

```python
# Baixa a primeira seção dos primeiros 300 mil artigos da wikipedia (não são os artigos completos).
from google_drive_downloader import GoogleDriveDownloader as gdd

gdd.download_file_from_google_drive(file_id='1QXY7Nh84FEtXVCAxJ3JhMaB_yfJGvYNH', dest_path='./wiki.pt-br_part.txt', unzip=False)
```

```python
# Abre o arquivo dos textos como minusculo
file = open("wiki.pt-br_part.txt","r") 
text = file.read().lower()
```

```python
# Quebra as seções em uma lista ( o \n é o separador entre as seções )
raw = text.split("\n")
print(raw[3])
tokens_sentenca = []
for sentenca in raw:
  # Pra cada seção, quebra ela em palavras e adiciona numa lista
  tokens_sentenca.append(sentenca.split())

# Aqui contem uma lista das seções quebrada em palavras
print(tokens_sentenca[:5])
```

```python
model = Word2Vec(tokens_sentenca, size=400, window=5, min_count=5, workers=multiprocessing.cpu_count())
```

* **init_sims(replace=True)**: Normaliza o modelo para não demandar tanta memória.

```python
# trim unneeded model memory = use (much) less RAM
model.init_sims(replace=True)
```

```python
# Salva o modelo em disco
model_name = "wiki.pt-br.word2vec.model"
model.save(model_name)
```
